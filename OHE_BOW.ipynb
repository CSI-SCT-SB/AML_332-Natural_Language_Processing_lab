```python
import nltk
import re
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import LancasterStemmer
from collections import Counter

text= input("Enter contents of the document: ")

sentence= sent_tokenize(text)
sentences= []

for sent in sentence:
    sent= re.sub(r'[.!,*)@;#%(&$_?^]', '', sent)
    sent= re.sub(r"[\n\t]", '', sent)
    sent= re.sub("[-]", ' ', sent)
    sent= re.sub("<.*?>", ' ', sent)
    sentences.append(sent)

stemmer = LancasterStemmer()

stop_words= set(stopwords.words('english'))
final_sent= []
for sent in sentences:
    temp= []
    tokens= word_tokenize(sent)
    for word in tokens:
        if word not in stop_words:
            stemmedword= stemmer.stem(word)
            temp.append(stemmedword)
    final_sent.append(temp)
    
print("Initial Document: \n", text)

print("Document after preprocessing: \n", final_sent)

vocab= [word for sent in final_sent for word in sent]
tokendict= Counter(vocab)

counts= [value for value in tokendict.values()]
counts.sort()

topdict= {}
for key, value in tokendict.items():
    if value in counts:
        topdict[key]= value
vocab= [key for key in topdict.keys()]
print("Words: \n", vocab)

table= []
for sent in final_sent:
    vector= []
    for word in vocab:
        if word in sent:
            vector.append(1)
        else:
            vector.append(0)
    table.append(vector)
print("\nBag of Words:")
i= 0
for row in table:
    print(sentences[i])
    i+= 1
    print(row)

table= []
i,j= 0,0
for sent in final_sent:
    ohe_vector= []
    for currentword in sent:
        word_vector= [0 for i in range(len(vocab))]
        for word in vocab:
            i+= 1
            if currentword == word:
                j+= 1
                word_vector[vocab.index(word)]= 1
        ohe_vector.append(word_vector)
    table.append(ohe_vector)
    
print("\nOne Hot Encoding:")
i= 0
for row in table:
    print(sentences[i])
    i+= 1
    print(row)
```
